```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score, accuracy_score, precision_score, f1_score, recall_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Crop_recommendation.csv')
```

```python
# detecting outliers 

crop_features = df.drop('label', axis = 1)

def detect_outliers(crop_features):
    outliers = pd.DataFrame(index = crop_features.columns)

    Q1 = crop_features.quantile(0.25)
    Q3 = crop_features.quantile(0.75)
    IQR = Q3 - Q1

    outliers['lower bound'] = Q1 - 1.5*IQR
    outliers['upper bound'] = Q3 + 1.5*IQR

    outliers['Count of Outliers'] = [
        ((crop_features[col] < outliers.loc[col, 'lower bound']) |
          (crop_features[col] > outliers.loc[col, 'upper bound'])).sum()
        for col in crop_features.columns
    ]
    return outliers
print("Outlier Analysis: \n")
print(detect_outliers(crop_features))
```

```python
# Visualize distributions
fig, axes = plt.subplots(2, 4, figsize=(12,6))

for ax, col in zip(axes.flat[:len(crop_features.columns)], crop_features.columns): 
    # Create a boxplot
    sns.boxplot(data=crop_features, y=col, ax=ax)
axes.flat[-1].remove()
fig.suptitle('Feature Distributions and Outlier Patterns', y=1)
plt.tight_layout(h_pad=1, w_pad=5)  # Increase spacing between subplots
plt.show()
```

```python
X = df.drop('label', axis = 1)

# label encoding all the label values 
le = LabelEncoder()
y = le.fit_transform(df['label'])
```

```python
print(f"len(X): {X.shape}, len(y): {y.shape}")
```

```python
# splitting the data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234)
```

```python
print(f"X_train: {X_train.shape} \n y_train: {y_train.shape} \n X_test: {X_test.shape} \n y_test: {y_test.shape}")
```

```python
pipeline = Pipeline([
    ('scaler', RobustScaler()),  # Scale data
    ('classifier', LogisticRegression())  # Placeholder, will be replaced in GridSearchCV
])

# Define hyperparameters for different models
param_grid = [
    {
        'classifier': [LogisticRegression()],
        'classifier__C': [0.1, 1, 10]
    },
    {
        'classifier': [RandomForestClassifier()],
        'classifier__n_estimators': [50, 100, 150],
        'classifier__max_depth': [None, 5, 10]
    },
    {
        'classifier': [DecisionTreeClassifier()],
        'classifier__max_depth': [None, 5, 10],
        'classifier__min_samples_split': [2, 5, 10]
    },
    {
        'classifier': [SVC()],
        'classifier__C': [0.1, 1, 10],
        'classifier__kernel': ['linear', 'rbf']
    }
]
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, verbose=2, scoring = 'f1')
grid_search.fit(X_train, y_train)
```

```python
print(f"Best Model: {grid_search.best_estimator_}")
print(f"Best Parametes: {grid_search.best_params_}")
```

```python
classifiers = [LogisticRegression, RandomForestClassifier, DecisionTreeClassifier, SVC]

best_params_per_model = {}

for params in grid_search.cv_results_['params']:
    model_name = params['classifier'].__class__.__name__
    if model_name not in best_params_per_model:
        best_params_per_model[model_name] = params

for model_name, best_params in best_params_per_model.items():
    model = Pipeline([
        ('scaler', RobustScaler()),
        ('classifier', best_params['classifier'])
    ])
    
    model.set_params(**best_params)
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    print(f"Model: {model_name}")
    print(classification_report(y_test, y_pred))
    print("-" * 80)
```

## Logistic Regression is a better model overall as it requires less computational power and has almost the same results as randomforest 

```python
best_model = LogisticRegression(C = 10)
best_model.fit(X_train, y_train)
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
prediction_model = pd.DataFrame({
    "Actual Value": le.inverse_transform(y_test),
    "Predicted Value": le.inverse_transform(y_pred)
})
```
