```python
def calculate_average_crop_data(df):
    # (Function code remains the same as in the previous response)
    df = df.rename(columns={
        'RICE PRODUCTION (1000 tons)': 'Rice_Production',
        'RICE YIELD (Kg per ha)': 'Rice_Yield',
        'WHEAT AREA (1000 ha)': 'Wheat_Area',
        'WHEAT PRODUCTION (1000 tons)': 'Wheat_Production',
        'WHEAT YIELD (Kg per ha)': 'Wheat_Yield',
        'KHARIF SORGHUM AREA (1000 ha)': 'Kharif_Sorghum_Area',
        'KHARIF SORGHUM PRODUCTION (1000 tons)': 'Kharif_Sorghum_Production',
        'KHARIF SORGHUM YIELD (Kg per ha)': 'Kharif_Sorghum_Yield',
        'SORGHUM AREA (1000 ha)': 'Sorghum_Area'
    })

    crop_columns = [
        ('Rice', ['Rice_Area', 'Rice_Production', 'Rice_Yield']),
        ('Wheat', ['Wheat_Area', 'Wheat_Production', 'Wheat_Yield']),
        ('Kharif Sorghum', ['Kharif_Sorghum_Area', 'Kharif_Sorghum_Production', 'Kharif_Sorghum_Yield']),
        ('Sorghum', ['Sorghum_Area'])
    ]

    results = []

    for crop, cols in crop_columns:
        temp_df = df[['District'] + cols].copy()
        temp_df['Crop'] = crop

        if len(cols) == 3:
            temp_df = temp_df.rename(columns={cols[0]: 'Area', cols[1]: 'Production', cols[2]: 'Yield'})
        elif len(cols) == 1:
            temp_df = temp_df.rename(columns={cols[0]: 'Area'})

        results.append(temp_df)

    combined_df = pd.concat(results, ignore_index=True)

    if 'Production' in combined_df.columns and 'Yield' in combined_df.columns:
        average_crop_data = combined_df.groupby(['District', 'Crop'])[['Area', 'Production', 'Yield']].mean()
    else:
        average_crop_data = combined_df.groupby(['District', 'Crop'])[['Area']].mean()

    return average_crop_data

# Example Data (simulated from your image)
data = """
District,RICE PRODUCTION (1000 tons),RICE YIELD (Kg per ha),WHEAT AREA (1000 ha),WHEAT PRODUCTION (1000 tons),WHEAT YIELD (Kg per ha),KHARIF SORGHUM AREA (1000 ha),KHARIF SORGHUM PRODUCTION (1000 tons),KHARIF SORGHUM YIELD (Kg per ha),SORGHUM AREA (1000 ha)
Gurugram,23,1544.72,135.7,188,1385.41,55.2,10,181.16,55.2
Gurugram,21,1615.38,160.3,274,1709.3,39.4,7,203.05,39.4
Gurugram,, , , , , , , ,238
"""

# Create a DataFrame from the example data
df = pd.read_csv(StringIO(data))

# Calculate the average crop data
average_data = calculate_average_crop_data(df)

# Print the result
print(average_data)
```
